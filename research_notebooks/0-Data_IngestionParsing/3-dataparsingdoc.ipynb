# Word Document processing

```python
# import python-docx
from docx import Document as DocxDocument
import os
from langchain_community.document_loaders import Docx2txtLoader,UnstructuredWordDocumentLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
## Mehod1: Using DocxtxtLoader
print("Method 1: Using Docx2txtLoader")
try:
    docx_loader = Docx2txtLoader("data/word_doc/proposal.docx")
    docs=docx_loader.load()
    print(f"Number of documents loaded: {len(docs)}")
    print(f"Document metadata: {docs[0].metadata}")
    print(f"First document content:\n{docs[0].page_content[:500]}...")
except Exception as e:
    print(f"Error loading document with Docx2txtLoader: {e}")
      
```

```python
## method 2
print("\nMethod 2: Using UnstructuredWordDocumentLoader")
try:
    unstructured_loader = UnstructuredWordDocumentLoader("data/word_doc/proposal.docx",model="elements")
    unstructured_docs=unstructured_loader.load()
    print(f"Number of documents loaded: {len(unstructured_docs)}")
    for i, doc in enumerate(unstructured_docs[:3]):
        print(f"\n--- Document {i+1} ---")
        print(f"Document metadata: {doc.metadata.get('category', 'unknown')}")
        print(f"First document content:\n{doc.page_content[:100]}...")
except Exception as e:
    print(f"Error loading document with UnstructuredWordDocumentLoader: {e}")
```
